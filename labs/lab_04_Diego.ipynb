## Problema 01

En la carpeta data se encuentra el archivo `nba.db`, el cual muestra informacion básica de algunos jugadores de la NBA.

```python
from sqlalchemy import create_engine
import pandas as pd
import os  

# Crear conector
conn = create_engine(os.path.join('sqlite:///','data', 'nba.db'))
```

```python
# funcion de consultas
def consulta(query,conn):
    return pd.read_sql_query(query, con=conn)
```

```python
# ejemplo
query = """
SELECT * FROM player
"""
consulta(query,conn)
```

El objetivo es tratar de obtener la mayor información posible de este conjunto de datos mediante código **SQL**. Para cumplir con este objetivo debe resolver las siguientes problemáticas:

1. Mostrar las primeras 5 filas

```python
# operador LIMIT, este operador nos permite limitar la cantidad de filas que se muestran por pantalla.
query = """
SELECT * FROM player
LIMIT 5
"""
consulta(query,conn)
```

2. Seleccionar los valores únicos de la columna `position`.

```python
# operador SELECT, este operador nos permite elegir la columna a mostrar por pantalla.

query = """
SELECT 
DISTINCT position
FROM player
"""
consulta(query,conn)
```

3. Seleccionar y cambiar el nombre de la columna `name` por `nombre` 

```python
# operador as, este operador nos permite modificar el nombre de la columna que se muestran por pantalla.

query = """
SELECT name as nombre FROM player
"""
consulta(query,conn)
```

4. Determinar el tiempo (en años) de cada jugador en su posición

```python
# Se crea una nueva columna nombrada (number_of_years) que se obtiene restando el año de inicio y fin de la carrera del jugador
# Se adiciona +1 para compensar el año que se pierde en la resta, es decir, incluyo los años extremos.

query = """
SELECT name, (year_end-year_start+1) AS number_of_years FROM player
"""
consulta(query,conn)
```

5. Encontrar el valor máximo de la columna `weight` por cada  valor de la columna `position`

```python
# Se utiliza el operador GROUP BY que nos permite agrupar según nosotros definamos.
# En esta ocasión se agrupo por "position" y se muestra por pantalla el peso máximo nombrado como "Max_Weight"
query = """
SELECT name, position, MAX(weight) as Max_Weight FROM player GROUP BY position
"""
consulta(query,conn)
```

6. Encontrar el total de jugadores por cada valor de la columna `year_start`

```python
# Se utiliza el comando GROUP BY para agrupar según año de inicio.
# Se utiliza el comando COUNT() para contar la cantidad de veces que aparece el año en cuestión.
# Adicionalmente, se muestra el año de inicio contado y la cantidad de veces que se repite
query = """
SELECT year_start, COUNT(year_start) as Number_of_players FROM player GROUP BY year_start
"""
consulta(query,conn)
```

7. Encontrar el valor mínimo, máximo y promedio de la columna `weight` por cada valor de la columnas `college`

```python
# Se utilizan operadores matemáticos (MIN, MAX, AVG) junto con GROUP BY
# De esta forma una vez agrupados según "college", se aplican los operadores matemáticos.
query = """
SELECT college, MIN(weight),MAX(weight),AVG(weight) FROM player GROUP BY college
"""

consulta(query,conn)
```

8. Filtrar por aquellos jugadores que cumplan con :
* Para la columna `year_start` tienen un valor mayor 1990 y menor a 2000
* Para la columna `position` tienen un valor de `C`,`C-F` o `F-C`
* Para la columna `college` tienen un valor distinto de `Duke University`


**Se entiende para esta pregunta el aplicar todas las condiciones al mismo tiempo, no de forma individual**

```python
# Los operadores > y <, se aplican para obtener a quienes tienen como año de inicio un valor mayor 1990 y menor a 2000
# No se utiliza el operador BETWEEN ya que este incluye los años en cuestión y se necesita un mayor/menor estricto.
# Para la positición se utiliza el operador "=", otra opción era el operador "LIKE".
# Para college se utilizó el operador diferente a ("<>").
# el AND y el OR son operadores lógicos  los cuales permiten formar expresiones booleanas tomando como operandos otras 
# expresiones booleanas
query = """
SELECT *  FROM player 
WHERE (position = 'C' OR position = 'C-F' OR position = 'F-C')
    AND college <> 'Duke University'AND year_start>1990 AND year_start<2000
"""

consulta(query,conn)
```

9. Crear dos conjuntos de datos y juntarlos en una misma *query*. Las condiciones de cada uno de los cojunto de datos son:

* **df1**:
    * Para la columna `year_start` tienen un valor mayor 1990 y menor a 2000
    * Para la columna `position` tienen un valor de `C`,`C-F` o `F-C`
   
* **df2**:
    * Para la columna `year_end` tienen un valor menor a 2000
    * Para la columna `position` tienen un valor de `G`o `F` 

```python
# Para la positición se utiliza el operador "="
# Los operadores > y <, se aplican para obtener a quienes tienen como año de inicio un valor mayor o menor según corresponda.

# Se crearon los dos conjuntos de datos basados en las condiciones indicadas en el enunciado
df1 = """
SELECT *  FROM player 
WHERE (position = 'C' OR position = 'C-F' OR position = 'F-C')
    AND year_start>1990 AND year_start<2000
"""

df2 = """
SELECT *  FROM player 
WHERE (position = 'G' OR position = 'F')
    AND year_end<2000
"""

# Mediante el uso del operador UNION se juntan ambos conjuntos de datos en un mismo query.
query = df1 + 'UNION' + df2
print(query)
consulta(query,conn)
```

## A continuación, se crear query distintos en base a lo solicitado
En el último apartado se realiza la "UNION" en una sola línea sin generar bases de datos distintas.

```python
# El operador UNION se utiliza para combinar el conjunto de resultados de dos o más selecciones SELECT.
# Para la positición se utiliza el operador "="
# Los operadores > y <, se aplican para obtener a quienes tienen como año de inicio un valor mayor o menor según corresponda.

query = """
SELECT *  FROM player 
WHERE (position = 'C' OR position = 'C-F' OR position = 'F-C')
    AND year_start>1990 AND year_start<2000
"""

df1 = consulta(query,conn)
df1
```
